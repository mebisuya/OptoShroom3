# Graph analysis of Single Cell Stimulation project


This code is made to analyze all the data stored from segmentation of MDCK cells expressing OptoShroom and stimulated with 488nm light

```python
import numpy as np
import matplotlib.pyplot as plt
from os.path import join
from os import listdir
# Set matplotlib backend
%matplotlib inline 
```

```python
DataOutputFolder = r"\\ebisuya.embl.es\ebisuya\Shared\Invagination Project\Single Cell Stimulation Project\OutputMeasurements"
FolderToSave = r"\\ebisuya.embl.es\ebisuya\Shared\Invagination Project\Single Cell Stimulation Project\TotalGraphs"


listofnames = listdir(DataOutputFolder)
print(listofnames)
NumbCells = len(listofnames)

listoffiles = []
for i in range(len(listofnames)):
    loadfile = join(DataOutputFolder, listofnames[i])
    temp = np.genfromtxt(loadfile, delimiter=',')
    listoffiles.append(temp)
print(NumbCells)
print(listoffiles[15])


DataOutputFolderCont = r"\\ebisuya.embl.es\ebisuya\Shared\Invagination Project\Single Cell Stimulation Project\OutputMeasurementsControls"


listofnamescont = listdir(DataOutputFolderCont)
print(listofnamescont)
NumbCellsCont = len(listofnamescont)

listoffilescont = []
for i in range(len(listofnamescont)):
    loadfile = join(DataOutputFolderCont, listofnamescont[i])
    temp = np.genfromtxt(loadfile, delimiter=',')
    listoffilescont.append(temp)
print(NumbCellsCont)
print(listoffilescont[9])
```

```python
#Now that we have all files we just have to create list with relative sizes:
RelativeSizes = []
for i in range(NumbCells):
    templist = listoffiles[i][:,1]/listoffiles[i][5,1] # Relative to the 6th timepoint, just before stim
    RelativeSizes.append(templist)

RelativeSizesCont = []
for i in range(NumbCellsCont):
    templist = listoffilescont[i][:,1]/listoffilescont[i][5,1] # Relative to the 6th timepoint, just before stim
    RelativeSizesCont.append(templist)
```

```python
#Now we can get mean and sd! 
#To get them we will need to put all timepoints together in the same list!: 
time = listoffiles[0].shape[0]

means = []
standDev = []
for t in range(time):
    temptimepoint = []
    for i in range(NumbCells):
        temptimepoint.append(RelativeSizes[i][t])
    tempmean = np.mean(temptimepoint)
    tempstd = np.std(temptimepoint)
    means.append(tempmean)
    standDev.append(tempstd)
print(means)
print(standDev)


meansCont = []
standDevCont = []
for t in range(time):
    temptimepoint = []
    for i in range(NumbCellsCont):
        temptimepoint.append(RelativeSizesCont[i][t])
    tempmean = np.mean(temptimepoint)
    tempstd = np.std(temptimepoint)
    meansCont.append(tempmean)
    standDevCont.append(tempstd)
print(meansCont)
print(standDevCont)
```

```python
#Now we can plot!! 
plt.figure(figsize = (12, 7), dpi=200)
for i in range(5, 55):
    plt.axvspan(i + 0.333, i + 0.917, color = '#78e3d8', alpha = 0.3)
for n in range(NumbCells):
    plt.plot(listoffiles[n][:,0], RelativeSizes[n], 'go--', linewidth = 3, markersize = 8, color = 'gray', alpha = 0.2)
for n in range(NumbCellsCont):
    plt.plot(listoffilescont[n][:,0], RelativeSizesCont[n], 'go--', linewidth = 3, markersize = 8, color = 'red', alpha = 0.2)
plt.plot(listoffilescont[0][:,0], meansCont, 'go--', linewidth = 3, markersize = 8, color = 'red', label = 'mean Control' )
plt.plot(listoffiles[0][:,0], means, 'go--', linewidth = 3, markersize = 8, color = 'black', label = 'mean' )
plt.title('Apical Surface Measurements', fontsize = 20)
plt.xlabel('Time (minutes)', fontsize = 15)
plt.ylabel('Surface Area Percentages', fontsize = 15)
ax = plt.axes()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.legend(fontsize = '18')
plt.savefig(filesaver +'_ConstrictionGraph.png',  bbox_inches='tight')
plt.show()
```
